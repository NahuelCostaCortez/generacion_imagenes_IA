## Librerías

Las librerías son una manera de organizar las diferentes versiones de código de forma sencilla de tal manera que instalando la librería no tenemos que descargar el repositorio, si no que se instala de forma global en el sistema para utilizarla en cualquier proyecto.


Para instalar una librería en Python se utiliza el siguiente comando:



```
# pip install nombre-librería
```



Ejemplo

````python
!pip install textblob
````

````python
from textblob import TextBlob

# Frase a analizar
es_text = "Me encanta escuchar música"

# Codificar y traducir a inglés
text = TextBlob(es_text).translate(from_lang='es', to='en')

# Analizar la polaridad de las palabras
sentiment = text.sentiment.polarity

if sentiment > 0.0:
	print('👍 Positivo: ', sentiment)
else:
	print('👎 Negativo: ', sentiment)
````
